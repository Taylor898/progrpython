```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', encoding = "ISO-8859-1")
df.sample()
```

```python
sns.lineplot(x = df['sqft_living'], y = df['bathrooms'])
plt.xlabel('N of bedrooms')
plt.ylabel('N of bathrooms')
plt.title('N of bedrooms/Bathrooms dependence')
```

### Выводы
На данном графике наблюдается прямая зависимость между площадью недвижимости количеством ванных комнат, из чего можно сделать предположение, что количество комнат/количество ванных будет иметь схожий график

```python
sns.lineplot(x = df['bedrooms'], y = df['bathrooms'])
plt.xlabel('N of bedrooms')
plt.ylabel('N of bathrooms')
plt.title('N of bedrooms/Bathrooms dependence')
```

### Выводы
Данный график опроверг выдвинутую ранее теорию, но создал новую: рост количества комнат связан с необходимостью разместить обслуживающий персонал на территории дома, при размещении которого можно исходить из одной ванной на несколько коек
